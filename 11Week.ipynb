#### 6. 파생변수 추가
###### df.assing()에 새 변수명 = 공식

```python
exam.assign(
    total = exam['math'] + exam['english'] + exam['science']
).head()
```

###### 여러 파생변수 한 번에 추가하기

```python
exam.assign(
    total = exam['math'] + exam['english'] + exam['science'],
    mean = (exam['math'] + exam['english'] + exam['science'])/3
).head()
```

###### df.assign()에 np.where() 적용

```python
import numpy as np
exam.assign(
test = np.where(exam['science'] >= 60, 'pass', 'fail')
)
```

###### 추가한 변수를 pandas 함수 .sort_values() 활용

```python
exam.assign(
total = exam['math'] + exam['english'] + exam['science']
).sort_values('total', ascending = True) #오름차순
```

```python
exam.assign(
total = exam['math'] + exam['english'] + exam['science']
).sort_values('total', ascending = False) #내림차순
```

#### 7. lambda 활용

###### lambda를 활용해서 데이터 프레임명을 대체할 수 있음

```python
loooooongName = pd.read_csv('exam.csv')
loooooongName.head()
```

```python
# 긴 데이터 프레임을 직접 입력
loooooongName.assign(
new = loooooongName['math'] + loooooongName['english'] + loooooongName['science']
).head()
```

```python
# lambda 활용

loooooongName.assign(
new = lambda x : x['math'] + x['english'] + x['science']
).head()
```

###### 본론으로 돌아와 다시 lambda를 활용해 total과 mean을 간결하게 작성

```python
exam.assign(
    total = lambda x : x['math'] + x['english'] + x['science'],
    mean = lambda x : x['total'] / 3
).head()
```

###### 파생 변수를 이용해 다시 파생 변수를 만들 때는 lambda를 이용하지 않고 직접 데이터 프레임명을 입력하면 KeyError 또는 Define 오류뜸

```python
exam.assign(
    total = exam['math'] + exam['english'] + exam['science'],
    mean = exam['total'] / 3
).head()
```
